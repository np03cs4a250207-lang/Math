```python
# Task - 4 (f)
import matplotlib.pyplot as plt
x = ["Red", "Silver","Blue", "White", "Green"]
color = ["Red", "Silver","Blue", "Purple", "Green"]
y = [123, 230, 78, 193, 12]
myexplode = [0,0.05, 0, 0, 0]
plt.subplot(2, 2, 1)
plt.pie(y, labels=x, colors=color, explode=myexplode)
plt.subplot(2, 2, 2)
plt.pie(y, labels=x, colors=color,explode=myexplode,shadow=True)
plt.subplot(2, 2, 3)
plt.pie(y, labels=x, colors=color)
plt.legend(title="Car Colours")
plt.subplot(2, 2, 4)
plt.pie(y, labels=x, colors=color)
plt.show()
```
